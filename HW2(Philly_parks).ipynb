```python
# Calculate the mean and median for acreage
mean_acreage = pparks['ACREAGE'].mean()
median_acreage = pparks['ACREAGE'].median()

# Print the results
print
print(f'Mean acreage: {mean_acreage}')
print(f'Median acreage: {median_acreage}')
```

The Mean value is much higher than median value. This indicates that there are some large parks with much larger acreage than most others, which skews the mean upward.

The median area is about 3.6, suggesting that half of the parks have an area smaller than this value, while the other half are larger. It also shows that this is a typical length for the parks.

Overall, the data suggests a wide range of park acreage, with a small number of exceptionally large parks inflating the average values. The median values give a better sense of the "typical" park acreage.

```python
# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(pparks['ACREAGE'], bins=20, color='skyblue', edgecolor='black')
plt.axvline(mean_acreage, color='red', linestyle='dashed', linewidth=1.5, label=f'Mean: {mean_acreage:.2f}')
plt.axvline(median_acreage, color='green', linestyle='dashed', linewidth=1.5, label=f'Median: {median_acreage:.2f}')
plt.title('Histogram of Park Sizes')
plt.xlabel('Park Size in Acerage')
plt.ylabel('Number of Parks')
plt.grid(True)
plt.legend()
plt.show()
```

The figure shows the frequency of parks of different acerage as well as their median and mean values. The x-axis represents the park size in acreage, and the y-axis represents the number of parks.

It shows that the majority of parks have acerage between 0 and 250. There are small number of parks with high acerage such as between 1500 and 1750.

The Mean value is much higher than median value which shows that this distribution is right skewed. This indicates that there are some large parks with much larger acreage than most others, which skews the mean upward.

The median area is about 3.6, suggesting that half of the parks have an area smaller than this value, while the other half are larger. It also shows that this is a typical length for the parks.

Overall, the data suggests a wide range of park acreage, with a small number of exceptionally large parks inflating the average values. The median values give a better sense of the "typical" park acreage.

```python
#standard deviation
park_sd= pparks['ACREAGE'].std()

#printsd
print(f'park_standard deviation: {park_sd}')
```

```python
# Filter parks smaller than 10 acres
small_parks = pparks[pparks['ACREAGE'] < 10]
small_parks
```

```python
#mean and median of small parks
mean_sp= small_parks['ACREAGE'].mean ()
median_sp = small_parks['ACREAGE'].median ()

#Print
print(f'mean small parks: {mean_sp}')
print(f'median small parks: {median_sp}')
```

```python
# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(small_parks['ACREAGE'], bins=20, color='skyblue', edgecolor='black')
plt.axvline(mean_sp, color='red', linestyle='dashed', linewidth=1.5, label=f'Mean: {mean_sp:.2f}')
plt.axvline(median_sp, color='green', linestyle='dashed', linewidth=1.5, label=f'Median: {median_sp:.2f}')
plt.title('Histogram of Park Sizes')
plt.xlabel('Park Size in Acerage')
plt.ylabel('Number of Parks')
plt.grid(True)
plt.legend()
plt.show()
```

The figure displays park size and frequency along with mean and median values as vertical lines. The X label indicates park size in acreage and Y label shows the number of parks. A high number of parks fall between 0 and 2 acreage. The majority of park acreage is relatively small and the distribution skews to the right. The small difference between the mean and median shows that there are less values on the extreme end of the distribution.

 The median value denotes that half the parks are below 2.17 acerage and the rest above. The mean is slightly higher denoting a small number of parks being much higher than median. Overall, both mean and median values give us useful information about the spread of data.

```python
#group parks by parent name and sum
parent_park= pparks.groupby('PARENT_NAME').sum()
parent_park
```

```python
#plot histogram
plt.figure(figsize=(10,6))
plt.hist(parent_park['ACREAGE'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Parent Park Size')
plt.xlabel('Park Size in Acerage')
plt.ylabel('Number of Parent Parks')
plt.grid(True)
plt.show()
```

The figure displays park size and frequency along with mean and median values as vertical lines. The X label indicates park size in acreage and Y label shows the number of parks. The overwhelming majority of parent parks are btw 0 and 500 acreage. A small number of parks have much higher acreage. This stands in contrast to the earlier datframe where we saw that majority park size was btw 0 and 250. Addtionally, none of the parks in the previous data set crossed 2000 acreage mark. Both histograms are right-skewed, but the skewness is much more extreme in the second histogram due to the inclusion of a few very large parks (parent parks). In summary, while both datasets highlight the predominance of smaller parks, the second histogram shows a wider range, including very large parks that aren't represented in the first.
